```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10,10
MAXLENGTH = 3
import matplotlib
import math
import matplotlib.patches as mpatches
```

```python
def l(hl): return math.log(2)/hl
```

```python
x = np.arange(0, 10000., 1.)
```

```python
plt.plot(x, 88*np.exp(-l(37)*x),linewidth=3.0, label='cesium-137')
plt.plot(x, 0.009*np.exp(-l(1.6*10**(6))*x),linewidth=3.0, label='uranium-233')
axes = plt.gca()
axes.set_ylim([0,0.02])
plt.legend(loc=1,prop={'size':20})
plt.title('Activity of 1 gram of cesium-137 and uranium-233 versus time',fontsize=16)
plt.xlabel('Time (years)', fontsize=20)
plt.ylabel('Activity (Ci)', fontsize=20)
plt.savefig('csur.png')
```
